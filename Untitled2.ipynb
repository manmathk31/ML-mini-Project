```python
import xgboost

# IMPORTS
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
import joblib
import json
import os

from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import AdaBoostRegressor

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

from xgboost import XGBRegressor

# CREATE FOLDERS
os.makedirs('models', exist_ok=True)
os.makedirs('plots', exist_ok=True)

# LOAD DATA
X_train = np.load('data/X_train.npy')
X_test = np.load('data/X_test.npy')

y_train = np.load('data/y_train.npy')
y_test = np.load('data/y_test.npy')

print("X_train shape:", X_train.shape)

# LOAD FEATURE NAMES
with open('data/feature_names.pkl', 'rb') as f:
    feature_names = pickle.load(f)

print("Total Features:", len(feature_names))

# CREATE MODELS
rf = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

gb = GradientBoostingRegressor(
    n_estimators=100,
    learning_rate=0.1
)

ada = AdaBoostRegressor(
    n_estimators=100,
    learning_rate=0.1
)

xgb = XGBRegressor(
    n_estimators=100,
    learning_rate=0.1,
    random_state=42
)

# TRAIN MODELS
print("Training Random Forest...")
rf.fit(X_train, y_train)

print("Training Gradient Boosting...")
gb.fit(X_train, y_train)

print("Training AdaBoost...")
ada.fit(X_train, y_train)

print("Training XGBoost...")
xgb.fit(X_train, y_train)

print("Training Completed")

# PREDICTIONS
y_pred_rf = rf.predict(X_test)
y_pred_gb = gb.predict(X_test)
y_pred_ada = ada.predict(X_test)
y_pred_xgb = xgb.predict(X_test)

# METRICS FUNCTION
def get_metrics(y_true, y_pred, name):

    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    print("\n", name)
    print("MSE :", round(mse, 4))
    print("RMSE:", round(rmse, 4))
    print("MAE :", round(mae, 4))
    print("R2  :", round(r2, 4))

    return {
        "MSE": float(mse),
        "RMSE": float(rmse),
        "MAE": float(mae),
        "R2": float(r2)
    }

# CALCULATE METRICS
m_rf = get_metrics(y_test, y_pred_rf, "Random Forest")
m_gb = get_metrics(y_test, y_pred_gb, "Gradient Boosting")
m_ada = get_metrics(y_test, y_pred_ada, "AdaBoost")
m_xgb = get_metrics(y_test, y_pred_xgb, "XGBoost")

# SAVE METRICS JSON
ensemble_metrics = {
    "Random Forest": m_rf,
    "Gradient Boosting": m_gb,
    "AdaBoost": m_ada,
    "XGBoost": m_xgb
}

with open('ensemble_metrics.json', 'w') as f:
    json.dump(ensemble_metrics, f, indent=2)

print("\nensemble_metrics.json saved")

# FEATURE IMPORTANCE FUNCTION
def plot_importance(model, name, filename):

    importances = model.feature_importances_

    indices = np.argsort(importances)[::-1][:15]

    plt.figure(figsize=(12,6))

    plt.bar(
        range(15),
        importances[indices]
    )

    plt.xticks(
        range(15),
        [feature_names[i] for i in indices],
        rotation=45,
        ha='right'
    )

    plt.title(name)

    plt.tight_layout()

    plt.savefig(f'plots/{filename}')

    plt.show()

# CREATE PLOTS
plot_importance(
    rf,
    'Random Forest Feature Importance',
    'rf_feature_importance.png'
)

plot_importance(
    gb,
    'Gradient Boosting Feature Importance',
    'gb_feature_importance.png'
)

plot_importance(
    xgb,
    'XGBoost Feature Importance',
    'xgb_feature_importance.png'
)

# SAVE MODELS
joblib.dump(rf, 'models/random_forest.pkl')
joblib.dump(gb, 'models/gradient_boosting.pkl')
joblib.dump(ada, 'models/adaboost.pkl')
joblib.dump(xgb, 'models/xgboost.pkl')

print("\nModels Saved")

# FINAL COMPARISON TABLE
with open('regression_metrics.json', 'r') as f:
    reg_metrics = json.load(f)

with open('ensemble_metrics.json', 'r') as f:
    ens_metrics = json.load(f)

all_metrics = {
    **reg_metrics,
    **ens_metrics
}

df_compare = pd.DataFrame(all_metrics).T

df_compare = df_compare.round(4)

df_compare.index.name = 'Model'

df_compare = df_compare.sort_values(
    'R2',
    ascending=False
)

print("\nFINAL COMPARISON TABLE")
print(df_compare)

# SAVE CSV
df_compare.to_csv(
    'final_comparison_table.csv',
    index=True
)

print("\nfinal_comparison_table.csv saved")

# FINAL R2 GRAPH
plt.figure(figsize=(14,6))

colors = ['steelblue']*5 + ['darkorange']*4

plt.bar(
    df_compare.index,
    df_compare['R2'],
    color=colors,
    alpha=0.85
)

plt.xticks(rotation=35, ha='right')

plt.ylabel('R2 Score')

plt.title('All Models R2 Comparison')

plt.tight_layout()

plt.savefig(
    'plots/all_models_r2_comparison.png'
)

plt.show()

print("\nDONE EVERYTHING SUCCESSFULLY")
```
